```python
#Code to import necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from time import sleep
from random import randint

```

```python
#Code to remove any white space in text of attributes
def space_cull(string):

    word = string.split(" ")
    result = ""
    for elem in word:
        if not elem == '':
            result += str(elem) + ' '
    return result.strip()
```

```python
#Web Scraping

#Note: pages need to be updated to current number of pages (At time of writing this code it is 91)
pages=91
#The following lists will hold the attributes we want extracted for each movie
movie_list = []
rating_list= []
metascore_list = []
genre_list = []
blurb_list = []
votes_list = []


for page in range(0,pages):

  domain= "https://www.imdb.com/"

  #Note: (page*50+1) is used because each page on IMDb currently shows 50 titles. If this changes, update to number of movies per page
  url = domain+'search/title/?title_type=feature&year=2023-01-01,2023-12-31&start='+str(page*50+1)+'&ref_=adv_nxt'

  webpage = requests.get(url, stream=True)
  soup = BeautifulSoup(webpage.text, "lxml")

  movie_info = soup.findAll(name="div", class_="lister-item-content")

  #Sleep is used to wait between requests so as to not get blocked 
  sleep(randint(5,10))

  for x in movie_info : 
      #Scrape title
      movie_title=x.find(name="h3", class_="lister-item-header")
      if movie_title is None:
        movie_list.apend("No title")
      else:
        movie_list.append(space_cull(movie_title.get_text()).strip())

      #Scrape rating
      movie_rating=x.find(name='div', class_="inline-block ratings-imdb-rating")
      if movie_rating is None:
        rating_list.append("No rating")
      else:
        rating_list.append(space_cull(movie_rating.get_text()).strip())

      #Scrape metascore
      movie_metascore=x.find(name='div',class_="inline-block ratings-metascore")
      if movie_metascore is None:
        metascore_list.append("No score")
      else:
        metascore_list.append(space_cull(movie_metascore.get_text()).strip())

      #Scrape genre
      movie_genre=x.find(name='span', class_="genre")
      if movie_genre is None:
        genre_list.append("No genre")
      else:
        genre_list.append(movie_genre.get_text().strip())

      #Scrape blurb
      movie_blurb=x.findAll(name='p', class_="text-muted")
      if movie_blurb is None:
        blurb_list.append("No blurb")
      else:
        blurb_list.append(space_cull(movie_blurb[1].get_text()).strip())

      #Scrape num votes
      movie_num_votes=x.find(name='p', class_="sort-num_votes-visible")
      if movie_num_votes is None:
        votes_list.append("No votes")
      else:
        votes_list.append(movie_num_votes.get_text().strip())

  #Print page number to track progress of scraping
  print("page "+str(page)+" is finished")
```

```python
#Create and populate data frame
movie_info_2023 = pd.DataFrame({ "Movie Title": movie_list, "Rating" : rating_list, "Metascore": metascore_list,"Genre": genre_list, "Blurb": blurb_list, "Votes" : votes_list })
```
